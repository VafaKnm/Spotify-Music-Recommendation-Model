**As you see, there is no missing data but this file have not the song IDs; For content based recommendation or any other recommendation system we need to know IDs or name of the songs for recommending them to the user**

**So, I am going to read json files and convert them to pandas dataframes**

```python
import json

with open('../input/spotify-recommendation/good.json', 'r') as f:
    data = json.load(f)

good_df = pd.DataFrame(data['audio_features'])
good_df
```

**There is some unnecessary information in the data such URI, URL and duration of song that I think it's better to drop them!**

**On the other hand, we add a column called like to indicate whether the song is good or bad (Of course, according to the opinion of the data producer!)**

```python
good_df = good_df.drop(columns=['type','uri','track_href','analysis_url','duration_ms'])
good_df['like'] = 1
good_df
```

```python
with open('../input/spotify-recommendation/dislike.json', 'r') as f:
    data = json.load(f)

bad_df = pd.DataFrame(data['audio_features'])
bad_df
```

```python
bad_df = bad_df.drop(columns=['type','uri','track_href','analysis_url','duration_ms'])
bad_df['like'] = 0
bad_df
```

**And now we can merge two dataframes to one**

```python
df = good_df.merge(bad_df, how='outer')
df
```

**For content based filtering we're going to use cosine similarity; however for a fairer comparison, we can drop like column and get their similarity only through the characteristics of each song**

```python
temp_df = df.drop(columns=['id', 'like'])
temp_df
```
